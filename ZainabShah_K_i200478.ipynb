```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("ECG200_TRAIN.csv",delimiter="  ")
df
```

```python
# data separation based on whether there was a heart attack or not i.e. 1 or -1 
pos = df[df['-1.0000000e+00'] == 1.0]
neg = df[df['-1.0000000e+00'] == -1.0]
```

```python
# storing the number of rows
p = len(pos)
n= len(neg)
```

```python
# finding splitting ratio 

splitP = p//4
splitN = n//4

print(splitP, splitN)
```

```python
#concatenating dataframes according to their splitting ratio

df1 = pd.concat([pos[:splitP], neg[:splitN]])
df2 = pd.concat([pos[splitP:2*splitP], neg[splitN:2*splitN]])
df3 = pd.concat([pos[2*splitP:3*splitP], neg[2*splitN:3*splitN]])
df4 = pd.concat([pos[3*splitP:70], neg[3*splitN:31]])
```

```python
print(df1['-1.0000000e+00'].value_counts())
print(df2['-1.0000000e+00'].value_counts())
print(df3['-1.0000000e+00'].value_counts())
print(df4['-1.0000000e+00'].value_counts())
```

```python
from sklearn.utils import shuffle
from sklearn.ensemble import IsolationForest 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report 
from sklearn.model_selection import KFold
```

```python
df1 = shuffle(df1)
df2 = shuffle(df2)
df3 = shuffle(df3)
df4 = shuffle(df4)
```

```python
accN = []
x = pd.concat([df1, df2, df3], axis=0)
y = df4
X_train = x.drop('-1.0000000e+00',axis=1)
y_train = x["-1.0000000e+00"]
X_test = y.drop('-1.0000000e+00',axis=1)
y_test = y["-1.0000000e+00"]


c3 = IsolationForest(random_state = 0)
c3.fit(X_train, y_train)
p = c3.predict(X_test)
print ("Accuracy : " , accuracy_score(y_test,p)*100)  
print("Report : \n", classification_report(y_test, p))
print("F1 Score : ",f1_score(y_test, p, average='macro')*100)
accN.append(accuracy_score(y_test,p)*100)
```

```python
# Now using Sklearn

x = df.drop('-1.0000000e+00',axis=1)
y = df['-1.0000000e+00']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 39)


c3 = IsolationForest(random_state = 0)
c3.fit(X_train, y_train)
p = c3.predict(X_test)
print ("Accuracy : " , accuracy_score(y_test,p)*100)  
print("Report : \n", classification_report(y_test, p))
print("F1 Score : ",f1_score(y_test, p, average='macro')*100)
```

```python
#Applying kfold cross validation

accS = []
kf = KFold(n_splits = 4,shuffle = True)
for i, j in kf.split(x): 
    X_train = x.iloc[i]
    y_train = y.iloc[i]
    X_test  = x.iloc[j]
    y_test  = y.iloc[j] 
    

    c3 = IsolationForest(random_state = 0)
    c3.fit(X_train, y_train)
    p = c3.predict(X_test)

    print ("Accuracy : " , accuracy_score(y_test,p)*100)
    accS.append(accuracy_score(y_test,p)*100)
```

```python
# Finding the average accuracies of both approaches 

avgN = sum(accN)/4
avgS = sum(accS)/4
```

```python
print("\nAverage Accuracies Using Numpy: ", avgN)
print("\nAverage Accuracies Using Sklearn: ", avgS)
```

As Numpy has higher level of abstractions it is performing slightly better than Sklearn
